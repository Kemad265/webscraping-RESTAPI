```python
import requests
from bs4 import BeautifulSoup
import sqlite3
```

```python
categories = {}
urls =[]
response = requests.request('GET', 'https://books.toscrape.com/')
link='https://books.toscrape.com/'
soup = BeautifulSoup(response.content, 'html.parser')
all_categ = soup.find('div', attrs={'class': 'side_categories'}).find('ul').find('li').find('ul').find_all('a')
for categ in all_categ:
    urls.append(link+categ.get('href'))

print (len(urls))
```

```python
rates = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
```

```python
link = 'https://books.toscrape.com/catalogue/'
connection=sqlite3.connect('books.db')
sql="INSERT INTO books (Name,Price,Rate,Description) VALUES (?,?,?,?) "
for url in urls:
    response = requests.request('GET', url)
    soup = BeautifulSoup(response.content, 'html.parser')
    all_books = soup.find_all('article', attrs={'class': 'product_pod'})
    for book in all_books:
        price = float(book.find('p', attrs={'class': 'price_color'}).get_text().split('£')[1])
        book_name = book.find('h3').find('a').get('title')
        rating = rates[book.find('p').get('class')[1]]
        response2 = requests.get(link+ book.findChildren('a')[1].attrs['href'][8:])
        soup2 = BeautifulSoup(response2.content)
        desc = soup2.find('meta', {'name': 'description'})['content']
        user = [book_name,price,rating,desc]
        connection.execute(sql,user)
        connection.commit()
        print(f'Book name: {book_name}, price: {price}, rating: {rating}\n')
        print("++++++++++++++++++++++++++++++++++")
        
connection.close()
```
